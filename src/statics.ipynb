# Clean words Count 

```python
clean_words_lable_0=0
clean_words_lable_1=0
clean_words_0=[]
clean_words_1=[]
with open('../data/dataset.csv', mode='r',encoding="utf8") as csv_file:
  csv_reader = csv.DictReader(csv_file)
  line_count = 0
  for row in csv_reader:
      if line_count == 0:
          print(f'Column names are {", ".join(row)}')
          line_count += 1
          continue
      # convert to lower case
      tokens = word_tokenize(row["lyrics"])
      #to lower
      tokens = [w.lower() for w in tokens]
      #remove non-alphabet
      table = str.maketrans('', '', string.punctuation)
      stripped = [w.translate(table) for w in tokens]
      words = [word for word in stripped if word.isalpha()]
      #filter out punct
      words = [word for word in tokens if word.isalpha()]
      #stop words
      stop_words = set(stopwords.words('english'))
      words = [w for w in words if not w in stop_words]
      if row["lable"]=="1":
        for w in words:
            clean_words_lable_1+=1
            clean_words_1.append(w)
      elif row["lable"]=="0":
        for w in words:
            clean_words_0.append(w)
            clean_words_lable_0+=1
      line_count += 1
  print(f'Processed {line_count} lines.')

print(clean_words_lable_0)
print(clean_words_lable_1)
```

```python
ax = plt.subplot(111, frame_on=False)
ax.xaxis.set_visible(False) 
ax.yaxis.set_visible(False)
plt.figure(figsize=(10,10)) 
rows=["Beafore 2000's","after 2000's"]
columns=["Counts of clean words"]
the_table = ax.table(cellText=[[clean_words_lable_0,clean_words_lable_1]],
                      rowLabels=["Counts of words"],
                      colLabels=rows,
                      loc='center')
fig1 = plt.gcf()
plt.savefig(Images_path+"Clean_words_table"+".png")
plt.show()
```

# TF-IDF


```python

unique0_count = dict.fromkeys(unique0,0)
unique1_count = dict.fromkeys(unique1,0)

clean_words_1
for w in clean_words_1:
    unique1_count[w]+=1
        
for w in clean_words_0:
    unique0_count[w]+=1 

TF0=dict.fromkeys(unique0_count,0.000)
TF1=dict.fromkeys(unique1_count,0.000)

for key, value in unique1_count.items():
    n=unique1_count[key]/len(clean_words_1)
    TF1[key]=n
for key, value in unique0_count.items():
    n=unique0_count[key]/len(clean_words_0)
    TF0[key]=n

IDF0=dict.fromkeys(unique0_count,0.000)
IDF1=dict.fromkeys(unique1_count,0.000)

for key, value in unique1_count.items():
    if key in common_count:
        IDF1[key]=math.log(2/2)
    else:
        IDF1[key]=math.log(2/1)
        
for key, value in unique0_count.items():
    if key in common_count:
        IDF0[key]=math.log(2/2)
    else:
        IDF0[key]=math.log(2/1)
    
TFIDF0=dict.fromkeys(unique0_count,0.000)
TFIDF1=dict.fromkeys(unique1_count,0.000)

for key, value in unique0_count.items():
    TFIDF0[key]=TF0[key]*IDF0[key]
for key, value in unique1_count.items():
    TFIDF1[key]=TF1[key]*IDF1[key]
        
TFIDF0=dict( sorted(TFIDF0.items(), key=operator.itemgetter(1),reverse=True))
TFIDF1=dict( sorted(TFIDF1.items(), key=operator.itemgetter(1),reverse=True))
```

```python
x0=[]
y0=[]
x1=[]
y1=[]
counter=0
for key, value in TFIDF0.items():
    x0.append(key)
    y0.append(value)
    counter+=1
    if counter==10:
        break
counter=0
for key, value in TFIDF1.items():
    x1.append(key)
    y1.append(value)
    counter+=1
    if counter==10:
        break
```
